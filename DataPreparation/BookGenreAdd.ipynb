```python
import pandas as pd
```

```python
book = pd.read_csv('books.csv')
tags = pd.read_csv('Bookstags.csv')
bookTags = pd.read_csv('book_tags.csv') 
```

```python
joint_tags=pd.merge(tags,bookTags,left_on='tag_id',right_on='tag_id',how='inner')
```

```python
joint_tags['tag_name'] = joint_tags['tag_name'].str.lower().str.strip()
```

```python
joint_tags
```

```python
p = joint_tags.groupby('tag_name').count()
p = p.sort_values(by='count', ascending=False)
```

```python
p
```

```python
genres = ["Art", "Biography", "Business", "Chick Lit", "Children's", "Christian", "Classics", "Comics", "Contemporary", "Cookbooks", "Crime", "Ebooks", "Fantasy", "Fiction", "Gay and Lesbian", "Graphic Novels", "Historical Fiction", "History", "Horror", "Humor and Comedy", "Manga", "Memoir", "Music", "Mystery", "Nonfiction", "Paranormal", "Philosophy", "Poetry", "Psychology", "Religion", "Romance", "Science", "Science Fiction", "Self Help", "Suspense", "Spirituality", "Sports", "Thriller", "Travel", "Young Adult"]
```

```python
genres = [genre.lower() for genre in genres]
```

```python
new_tags = p[p.index.isin(genres)]
```

```python
genres
```

```python
new_tags
```

```python
joint_tags['genre'] = joint_tags['tag_name'].apply(lambda x: x.lower() if x.lower() in genres else None)
```

```python
joint_tags[joint_tags['genre'].notnull()]
```

```python
fin=book.merge(joint_tags[joint_tags['genre'].notnull()], left_on='book_id', right_on="goodreads_book_id", how="inner")
```

```python
fin
```

```python
duplicate_rows = fin[fin.duplicated(subset='book_id', keep=False)]
#joint_tags_no_duplicates = fin.drop_duplicates(subset='original_title', keep='first')
```

```python
duplicate_rows
```

```python
fin['id'].dtype
```

```python
fin['genres'] = [None] * len(fin)
```

```python
count = 0
id_count = fin['id'].value_counts()

for i in range(1, 10001):
    genre = []
    
    for k in range(id_count.get(i, 0)):
        genre.append(fin.iloc[count]['genre'])
        count += 1
    
    fin.at[count-1, 'genres'] = genre
```

```python
fin
```

```python
duplicate_rows = fin.drop_duplicates(subset='original_title', keep='last')
```
